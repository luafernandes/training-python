### Gráfico Candlestick

Preparando os dados para a criação do **gráfico**

```python
df = dados.head(60).copy()
df['data'] =  df.index
df['data'] = df['data'].apply(mdates.date2num) #MDATES: método do matplotlib para manipulação de datas
```

```python
df
```

Desenhando o Gráfico de Velas

```python
fig, ax = plt.subplots(figsize=(15,8))

# Definindo a largura dos candles do gráfico
width = 0.7

#Inserindo a cor do candle
for i in range(len(df)):
    if df['fechamento'].iloc[i] > df['abertura'].iloc[i]:
        color='green'
    else:
        color='red'

    # Desenhando a linha vertical do candle (mecha)
    # Essa linha mostra os preços máximo (topo da linha) e mínimo (base da linha) do dia.
    # Usamos `ax.plot` para desenhar uma linha vertical.
    # [df['Data'].iloc[i], df['Data'].iloc[i]] define o ponto x da linha (a data), e [df['Mínimo'].iloc[i], df['Máximo'].iloc[i]] define a altura da linha.
    ax.plot([df['data'].iloc[i], df['data'].iloc[i]],
            [df['minimo'].iloc[i], df['maximo'].iloc[i]],
            color=color,
            linewidth=1)

    ax.add_patch(plt.Rectangle((df['data'].iloc[i] - width/2, min(df['abertura'].iloc[i], df['fechamento'].iloc[i])),
                                width,
                                abs(df['fechamento'].iloc[i]-df['abertura'].iloc[i]),
                                facecolor=color))

df['MA7'] = df['fechamento'].rolling(window=7).mean()
df['MA14'] = df['fechamento'].rolling(window=14).mean()

# Plotando as médias móveis
ax.plot(df['data'], df['MA7'], color='blue', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df['data'], df['MA14'], color='yellow', label='Média Móvel 14 Dias')  # Média de 14 dias
# Adicionando legendas para as médias móveis
ax.legend()

# Formatando o eixo x para mostrar as datas
# Configuramos o formato da data e a rotação para melhor legibilidade
ax.xaxis_date() #O método xaxis_date() é usado para dizer ao Matplotlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)

# Adicionando título e rótulos para os eixos x e y
plt.title("PETR4.SA - Gráfico de Candlestick")
plt.xlabel("Data")
plt.ylabel("Preço")

# Adicionando uma grade para facilitar a visualização dos valores
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

```python
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                    vertical_spacing=0.1,
                    subplot_titles=('Candlesticks', 'Volume Transacionado'),
                    row_width=[0.2, 0.7])

# Adicionando o gráfico de candlestick
fig.add_trace(go.Candlestick(x=df.index,
                             open=df['abertura'],
                             high=df['maximo'],
                             low=df['minimo'],
                             close=df['fechamento'],
                             name='Candlestick'),
                             row=1, col=1)

# Adicionando as médias móveis
fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA7'],
                         mode='lines',
                         name='MA7 - Média Móvel 7 Dias'),
                         row=1, col=1)

fig.add_trace(go.Scatter(x=df.index,
                         y=df['MA14'],
                         mode='lines',
                         name='MA14 - Média Móvel 14 Dias'),
                         row=1, col=1)

# Adicionando o gráfico de barras para o volume de transações, que nos dá uma ideia da atividade de negociação naquele dia
fig.add_trace(go.Bar(x=df.index,
                     y=df['volume'],
                     name='Volume'),
                     row=2, col=1)

# Atualizando layout
fig.update_layout(yaxis_title='Preço',
                  xaxis_rangeslider_visible=False,  # Desativa o range slider
                  width=1100, height=600)

# Mostrando o gráfico
fig.show()
```

* Buscar a ação da Apple e recriar o gráfico de Candlestick usando a biblioteca MPLFinance.

```python
dat_apple = yf.download('AAPL34.SA', start='2023-01-01', end='2023-12-31')
dat_apple
```

```python
# Define as cores personalizadas para o mercado
cores_mercado = mpf.make_marketcolors(up='green', down='red', edge='inherit', wick='inherit', volume='in', ohlc='i')

# Cria um estilo próprio
meu_estilo = mpf.make_mpf_style(marketcolors=cores_mercado,
                                gridstyle='-',
                                y_on_right=False,
                                gridaxis='both',
                                figcolor='white',  # Cor de fundo da figura
                                facecolor='lightgray')  # Cor de fundo do gráfico
```

```python
mpf.plot(dat_apple.head(50), type='candle', mav=(7,14), volume=True, style=meu_estilo, title="Meu Gráfico Candlestick", ylabel='Preço', ylabel_lower='Volume', figsize=(12, 6))
```

```python
# Verificando duplicatas no índice
print(dat_apple.index.is_unique)

# Dropando possíveis duplicatas
dat_apple = dat_apple[~dat_apple.index.duplicated(keep='first')]

# Verificando valores NaN
print(dat_apple.isna().sum())

# Removendo linhas com valores NaN, se necessário
dat_apple.dropna(inplace=True)
```

```python
# Calcular a Média Móvel Simples para 20 períodos
dat_apple['SMA'] = dat_apple['Close'].rolling(window=20).mean()

# Calcular o Desvio Padrão para 20 períodos
dat_apple['STD'] = dat_apple['Close'].rolling(window=20).std()

# Calcular Bandas de Bollinger
dat_apple['Upper'] = dat_apple['SMA'] + (dat_apple['STD'] * 2)
dat_apple['Lower'] = dat_apple['SMA'] - (dat_apple['STD'] * 2)


ap = [
    mpf.make_addplot(dat_apple['Upper'], color='green'),
    mpf.make_addplot(dat_apple['SMA'], color='blue'),
    mpf.make_addplot(dat_apple['Lower'], color='red')
]

# Plotar o gráfico
mpf.plot(dat_apple, type='candle', addplot=ap, style='charles', figratio=(12, 8), title="Apple Candlestick com Bandas de Bollinger")
```
