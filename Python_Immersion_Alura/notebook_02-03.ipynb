## Aula 03

Executar os mesmos passos que fizemos no Google Sheets, mas agora utilizando Python.

* Manipulando o DF Principal deixando as colunas que serão necessárias no momento

```python
df_principal = df_principal[['Ativo','Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

* Renomeando as colunas que possuem caracteres especiais

```python
df_principal = df_principal.rename(columns={'Último (R$)' : 'Valor_final', 'Var. Dia (%)': 'Var_dia'}).copy()
df_principal
```

* Criando as novas colunas que serão necessárias para as análises


```python
# Coluna variação do dia em percentual
df_principal['var_dia_pct'] = df_principal['Var_dia']/100

# Coluna para calcular o valor que a ação iniciou o dia
df_principal['valor_inicial_dia'] = df_principal['Valor_final'] / (df_principal['var_dia_pct'] + 1)

df_principal
```

* Criar uma nova coluna no df_principal que receberá a quantidade de ações de cada empresa, esse valor está contido no df_total_acoes.

Para trazer esses dados usaremos o MERGE

```python
df_principal = df_principal.merge(df_total_acoes, left_on = 'Ativo', right_on = 'Código', how = 'left')
df_principal
```

```python
df_principal = df_principal.drop('Código', axis = 1)
df_principal
```

* Criando a coluna que receberá o cálculo total da variação do dia em reais

```python
df_principal['variacao_RS'] = (df_principal['Valor_final'] - df_principal['valor_inicial_dia'])*df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format='{:.2f}'.format
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica' : 'Qtd_teorica'}).copy()
df_principal
```

* Criar a coluna que nos mostrará o resultado da empresa naquele dia (subiu, desceu ou estável)

Função APPLY: é chamada para aplicar outra função <br/>
Função LAMBDA: usada para operações linha a linha

```python
df_principal['Resultado'] = df_principal['variacao_RS'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal
```

* Criar uma nova coluna que receberá o nome da empresa, este nome está no df_ticker

```python
df_principal = df_principal.merge(df_ticker, left_on = 'Ativo', right_on = 'Ticker', how = 'left')
df_principal = df_principal.drop('Ticker', axis = 1)
df_principal
```

* Criar uma nova coluna que receberá o segmento e a idade da empresa, estes dados estão no df_segmento_idade

```python
df_principal = df_principal.merge(df_segmento_idade, left_on = 'Nome', right_on = 'Empresa', how = 'left')
df_principal = df_principal.drop('Empresa', axis = 1)
df_principal
```

* Criar uma nova coluna onde categorizaremos a idade das empresas em 3 faixas: <br/>
    - Menos de 50 anos
    - Entre 50 e 100 anos
    - Mai de 100 anos

```python
df_principal['Faixa_idade'] = df_principal['Idade em anos'].apply(lambda x: 'Mais de 100 anos' if x > 100 else ('Menos de 50 anos' if x < 50 else 'Entre 50 e 100 anos'))
df_principal
```

### Analisando os dados

```python
maior_valor = df_principal['variacao_RS'].max()
menor_valor = df_principal['variacao_RS'].min()
media_geral = df_principal['variacao_RS'].mean()
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['variacao_RS'].mean()
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['variacao_RS'].mean()

print(f"Maior:\t\tR$ {maior_valor:,.2f}")
print(f"Menor:\t\tR$ {menor_valor:,.2f}")
print(f"Média:\t\tR$ {media_geral:,.2f}")
print(f"Média (subiu):\tR$ {media_subiu:,.2f}")
print(f"Média (desceu):\tR$ {media_desceu:,.2f}")
```

- Quais foram as empresas com resultado positivo?

```python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_RS'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('Resultado')['variacao_RS'].sum().reset_index()
df_analise_saldo
```

### Criando os gráficos

Construir gráficos com a biblioteca Plotly Express

```python
import plotly.express as px
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='variacao_RS', text='variacao_RS', title='Variação R$ por Resultado')
fig.show()
```

```python
fig.update_traces(texttemplate='R$%{text:.2f}', textposition='outside')
fig.update_yaxes(tickprefix="R$", ticksuffix="", tickformat=",.")
fig.show()
```

```python
fig = px.pie(df_analise_segmento, values='variacao_RS', names='Segmento', title='Variação R$ por Segmento')
fig.update_layout(width=800,height=600)
fig.show()
```

```python
df_analise_idade = df_principal.groupby('Faixa_idade')['variacao_RS'].sum().reset_index()
fig = px.bar(df_analise_idade, x='Faixa_idade', y='variacao_RS', text='variacao_RS', title='Variação R$ por Faixa de Idade')
fig.update_traces(texttemplate='R$%{text:.2f}', textposition='outside')
fig.show()
```

```python
analise_idade = df_principal.groupby('Faixa_idade').agg({'variacao_RS': 'sum', 'Ativo': 'count'}).reset_index()
analise_idade.columns = ['Faixa_idade', 'Variação(R$)', 'Quantidade Empresas']
analise_idade['Variação(R$)'] = analise_idade['Variação(R$)'].map('${:,.2f}'.format)
df_analise_idade = pd.DataFrame(analise_idade)
df_analise_idade
```

```python
df_analise_idade['Variação(R$)'] = df_analise_idade['Variação(R$)'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
df_analise_idade['Variação(R$)'] = df_analise_idade['Variação(R$)'].astype(float)

df_analise_idade
```

```python
fig = px.bar(df_analise_idade, x='Faixa_idade', y='Variação(R$)', text='Variação(R$)',
             title='Variação de RS por Faixa de Idade',
             labels={'Variação(R$)': 'Variação (R$)', 'Faixa_idade': 'Faixa de Idade'})
fig.update_traces(texttemplate='R$%{text:.2f}', textposition='outside')
fig.show()
```
